Hay algunas funciones que se encuentran por defecto al momento de instalar R y hay otras que vamos aadiendo conforme instalamos paqueteras. Si queremos conocer qu paqueteras de R tenemos instaladas en nuestra computadora, utilizamos la funcin `installed.packages()`. Prueba usndolo en la siguiente celda para que veas qu paqueteras se encuentran instaladas justo en este momento.

```r
installed.packages()
```

En ocasiones habr momentos donde necesitemos hacer uso de funciones que no se encuentran en ninguna de las paqueteras preinstaladas (o en las que ya hemos instalado anteriormente). Intentemos ejecutar el comando `ggplot(data = iris, mapping = aes(x = Species, y = Sepal.Length)) + geom_boxplot()`

```r
ggplot(data = iris, mapping = aes(x = Species, y = Sepal.Length)) + geom_boxplot()
```

Observemos que nos marca error. Esto no es coincidencia, no nos dejar usar la lnea de cdigo anterior debido a que la funcin `ggplot()` no se encuentra en ninguna de las paqueteras que vimos al momento de correr la lnea de cdigo `installed.packages()`. Haciendo una bsqueda rpida en internet, encontramos que `ggplot()` se encuentra en la paquetera `ggplot2`.

Para instalar una paquetera en especfico, haremos uso de la funcin `install.packages()`. Instala la paquetera `ggplot2` con el comando `install.packages("ggplot2")` en la siguiente celda.

```r
install.packages("ggplot2")
```

Intentemos ejecutar nuevamente el comando `ggplot(data = iris, mapping = aes(x = Species, y = Sepal.Length)) + geom_boxplot()`. Qu sucede?

```r
ggplot(data = iris, mapping = aes(x = Species, y = Sepal.Length)) + geom_boxplot()
```

Ser que no fue instalada correctamente? Utilicemos el comando `is.element("ggplot2",installed.packages()[,1])` para verificar si fue instalada nuestra paquetera `ggplot2`.

Al momento de querer usar una funcin de una paquetera que no es parte de las funciones predefinidas no basta con instalar la paquetera que s la incluye. De hecho, slo es necesario instalarla una sola vez. Lo que s tenemos que hacer antes de usar una funcin en el entorno es usar la funcin `library()`. Ejecutar `library(ggplot2)` en la siguiente celda.

```r
library(ggplot2)
```

Ahora ejecutemos nuevamente el comando `ggplot(data = iris, mapping = aes(x = Species, y = Sepal.Length)) + geom_boxplot()`.

```r
ggplot(data = iris, mapping = aes(x = Species, y = Sepal.Length)) + geom_boxplot()
```
